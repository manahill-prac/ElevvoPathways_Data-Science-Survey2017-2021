```python
import pandas as pd

# Load the dataset
df = pd.read_csv(output)

# Display the first few rows
df.head()
```

```python
# Separate categorical and numeric columns
cat_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

# Fill categorical columns with 'Unknown'
for col in cat_cols:
    df[col] = df[col].fillna('Unknown')

# Fill numeric columns with median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Quick check
df.isnull().sum()
```

```python
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
print(f"Removed {initial_rows - df.shape[0]} duplicate rows.")
```

```python
from sklearn.preprocessing import LabelEncoder

# Standardize text: lowercase + strip spaces
for col in cat_cols:
    df[col] = df[col].str.strip().str.lower()

# Encode categorical columns
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
for col in cat_cols[:5]:  # show first 5 categorical columns
    print(f"Column: {col}")
    print("Unique values (first 10) after encoding:", df[col].unique()[:10])
    print("-"*50)
```

```python
# ---------- STEP 08:  Insights  ----------

# Example mapping (update with numeric codes if present)
country_map = {27:'United States', 75:'India', 46:'Germany', 10:'Brazil', 56:'Canada'}
gender_map = {1:'Male', 2:'Female', 3:'Other', 4:'Prefer not to say', 5:'Other'}
experience_map = {0:'<1 yr', 1:'1-2 yrs', 2:'2-3 yrs', 3:'3-4 yrs', 7:'5-10 yrs', 10:'10+ yrs'}
education_map = {0:'High School', 5:'Bachelors', 9:'Masters', 19:'PhD', 35:'Other'}

# Helper function: apply mapping only to numeric codes, leave text as is
def map_if_numeric(val, mapping):
    return mapping[val] if isinstance(val, (int, float)) and val in mapping else val

# Map columns
top_countries_readable = df['Q3'].apply(lambda x: map_if_numeric(x, country_map)).value_counts().head(5)
gender_dist_readable = df['Q2'].apply(lambda x: map_if_numeric(x, gender_map)).value_counts()
experience_dist_readable = df['Q6'].apply(lambda x: map_if_numeric(x, experience_map)).value_counts()
education_dist_readable = df['Q5'].apply(lambda x: map_if_numeric(x, education_map)).value_counts()

# Tools: already combined in previous step
tool_cols = [col for col in df.columns if col.startswith('Q7_Part')]
tools_data = df[tool_cols].apply(lambda x: x.value_counts()).fillna(0)
tools_sum = tools_data.sum(axis=1).sort_values(ascending=False).head(5)

# Display polished insights
print("📌 Top 5 Countries:\n", top_countries_readable)
print("\n📌 Gender Distribution:\n", gender_dist_readable)
print("\n📌 Experience Levels:\n", experience_dist_readable)
print("\n📌 Education Levels:\n", education_dist_readable)
print("\n📌 Top 5 Tools Used:\n", tools_sum)
```

```python
# Normalize country names
df['Q3'] = df['Q3'].replace({
    'United States of America': 'United States'
})

# Normalize gender labels
df['Q2'] = df['Q2'].replace({
    'Man':'Male', 'Woman':'Female',
    'Non-binary, genderqueer, or gender non-conforming':'Nonbinary'
})

# Normalize experience levels
df['Q6'] = df['Q6'].replace({
    '< 1 years':'<1 yr', '< 1 year':'<1 yr',
    '1 to 2 years':'1-2 yrs', '3 to 5 years':'3-5 yrs',
    '6 to 10 years':'5-10 yrs', 'More than 10 years':'10+ yrs'
})

# Recalculate distributions after cleaning
top_countries_readable = df['Q3'].value_counts().head(5)
gender_dist_readable = df['Q2'].value_counts()
experience_dist_readable = df['Q6'].value_counts()
education_dist_readable = df['Q5'].value_counts()
tools_sum = df[tool_cols].apply(lambda x: x.value_counts()).fillna(0).sum(axis=1).sort_values(ascending=False).head(5)

# Display cleaned polished insights
print("📌 Top 5 Countries:\n", top_countries_readable)
print("\n📌 Gender Distribution:\n", gender_dist_readable)
print("\n📌 Experience Levels:\n", experience_dist_readable)
print("\n📌 Education Levels:\n", education_dist_readable.head(15)) # top 15 only
print("\n📌 Top 5 Tools Used:\n", tools_sum)
```

```python
# ---------- STEP 09: Visual Insights ----------

import matplotlib.pyplot as plt
import seaborn as sns

# --- Top 5 Countries ---
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(
    x=top_countries_readable.index,
    y=top_countries_readable.values,
    palette='viridis',
    hue=top_countries_readable.index,  # assign x to hue
    dodge=False,
    legend=False
)
plt.title("📌 Top 5 Countries")
plt.ylabel("Count")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()


# --- Gender Distribution ---
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(
    x=gender_dist_readable.index,
    y=gender_dist_readable.values,
    palette='magma',
    hue=gender_dist_readable.index,
    dodge=False,
    legend=False
)
plt.title("📌 Gender Distribution")
plt.ylabel("Count")
plt.xlabel("Gender")
plt.xticks(rotation=45)
plt.show()


# --- Experience Levels ---
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(
    x=experience_dist_readable.index,
    y=experience_dist_readable.values,
    palette='coolwarm',
    hue=experience_dist_readable.index,
    dodge=False,
    legend=False
)
plt.title("📌 Experience Levels")
plt.ylabel("Count")
plt.xlabel("Experience")
plt.xticks(rotation=45)
plt.show()


# --- Education Levels (Top 10 for readability) ---
top_education = education_dist_readable.head(10)
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(
    x=top_education.index,
    y=top_education.values,
    palette='cividis',
    hue=top_education.index,
    dodge=False,
    legend=False
)
plt.title("📌 Education Levels (Top 10)")
plt.ylabel("Count")
plt.xlabel("Education / Role")
plt.xticks(rotation=45)
plt.show()


# --- Top 5 Tools Used ---
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(
    x=tools_sum.index,
    y=tools_sum.values,
    palette='rocket',
    hue=tools_sum.index,
    dodge=False,
    legend=False
)
plt.title("📌 Top 5 Tools Used")
plt.ylabel("Count")
plt.xlabel("Tool")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
from IPython.display import display

# Example: your cleaned dataset
df_readable = pd.read_csv("output/survey_2017_2021_readable.csv")

# Mappings
country_map = {27:'United States', 75:'India', 46:'Germany', 10:'Brazil', 56:'Canada'}
gender_map = {1:'Male', 2:'Female', 3:'Other', 4:'Prefer not to say', 5:'Other'}
experience_map = {0:'<1 yr', 1:'1-2 yrs', 2:'2-3 yrs', 3:'3-4 yrs', 7:'5-10 yrs', 10:'10+ yrs'}
education_map = {0:'High School', 5:'Bachelors', 9:'Masters', 19:'PhD', 35:'Other'}

# Helper
def map_if_numeric(x, mapping):
    return mapping.get(x, x)

# Prepare metrics
top_countries = df_readable['Q3'].apply(lambda x: map_if_numeric(x, country_map)).value_counts()
gender_dist = df_readable['Q2'].apply(lambda x: map_if_numeric(x, gender_map)).value_counts()
experience_dist = df_readable['Q6'].apply(lambda x: map_if_numeric(x, experience_map)).value_counts()
education_dist = df_readable['Q5'].apply(lambda x: map_if_numeric(x, education_map)).value_counts()

tool_cols = [col for col in df_readable.columns if col.startswith('Q7_Part')]
tools_data = df_readable[tool_cols].apply(lambda x: x.value_counts()).fillna(0)
top_tools = tools_data.sum(axis=1).sort_values(ascending=False)

# Maximum length for padding
max_len = max(len(top_countries), len(gender_dist), len(experience_dist), len(education_dist), len(top_tools))

# Helper to pad a Series
def pad_series(s, length):
    s = pd.Series(s)  # convert index/values to Series
    s = s.reindex(range(length), fill_value="")
    return s

# Combine into one summary dashboard
summary_df = pd.DataFrame({
    'Top 5 Countries': pad_series(top_countries.index, max_len),
    'Top Countries Count': pad_series(top_countries.values, max_len),
    'Gender Categories': pad_series(gender_dist.index, max_len),
    'Gender Count': pad_series(gender_dist.values, max_len),
    'Experience Categories': pad_series(experience_dist.index, max_len),
    'Experience Count': pad_series(experience_dist.values, max_len),
    'Education Categories': pad_series(education_dist.index, max_len),
    'Education Count': pad_series(education_dist.values, max_len),
    'Top Tools': pad_series(top_tools.index, max_len),
    'Top Tools Count': pad_series(top_tools.values, max_len)
})

# Display styled dashboard in Colab
print("📊 Survey Summary Dashboard")
display(
    summary_df.style.set_properties(**{
        'text-align': 'center',
        'font-size': '12pt'
    }).set_table_styles([{
        'selector': 'th',
        'props': [('text-align', 'center'), ('background-color', '#f2f2f2'), ('font-weight', 'bold')]
    }])
)
```
